```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import minisom
```

```python
with open('data1.txt', 'r') as f:
    data = np.loadtxt(f)

df = pd.DataFrame(data)
df
```

```python
plt.scatter(df[0], df[1])
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Дмаграмма рассеяния")
plt.show()
```

```python
som = minisom.MiniSom(2, 1, df.shape[1], sigma = 0.2, learning_rate = 0.5)
som.random_weights_init(df.values)
som.train_random(df.values, 50)
feature_map = som.win_map(df.values)
#feature_map
```

```python
labels = []

for cluster in df.values:
    labels.append(som.winner(cluster)[0])

plt.scatter(df[0], df[1], c = labels)

for center in som.get_weights():
    plt.scatter(center[:, 0], center[:, 1], marker = 'x', s = 100, c = 'red')

plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
